```python
# pip install h5py
import h5py
import numpy as np
import matplotlib.pyplot as plt

# Path to your HDF5 file
h5_file_path = 'R78H12-20190409_62_D3-m-40x-central-GAL4-unaligned_stack.h5j'

# Open the HDF5 file and extract image data as numpy array
with h5py.File(h5_file_path, 'r') as f:
    # Show all group keys for reference
    print("Keys: %s" % list(f.keys()))
    # Assuming image data is within f['Channels']['Channel_3']
    channel3 = f['Channels']['Channel_3']
    print("Shape of Channel_3:", channel3.shape)
    # Convert to numpy array
    img_array = np.array(channel3)
    print("Numpy array shape:", img_array.shape)
    print("Numpy array dtype:", img_array.dtype)
    # Optionally, display a slice of the array with matplotlib, e.g., first z stack if applicable
    if img_array.ndim == 3:  # (z, y, x)
        plt.imshow(img_array[0, :, :])
        plt.title("Channel_3, z=0")
        plt.show()
    elif img_array.ndim == 2:
        plt.imshow(img_array)
        plt.title("Channel_3")
        plt.show()
    else:
        print("Array shape not 2D or 3D, manual visualization needed")
```
